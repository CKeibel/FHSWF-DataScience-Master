# Univariate Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = 100
b = 200

def predict(x: np.ndarray, bias: float, weight: float) -> np.ndarray:
    pred = (w * x) + b
    for x_i, y_i in zip(x, pred):
        print(f"{w} * {x_i} + {b} = {y_i}")
    return pred
```

```python
x = np.arange(start=0, stop=8, step=1)
x
```

```python
y = np.linspace(start=150, stop=370, num=len(x))
y
```

```python
predict(x=x, weight=w, bias=b)
```

```python
plt.title("Data")
plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
plt.title("Data with Regression Line")
plt.scatter(x, y)
plt.plot(x, predict(x, bias=b, weight=w), color="red")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
w = 30
b = 150

plt.title("Data with Regression")
plt.scatter(x, y)
plt.plot(x, predict(x, bias=b, weight=w), color="red")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
